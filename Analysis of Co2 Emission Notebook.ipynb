## Important Column Description



>   `year`: This column represents the year in which the data was recorded. It allows for tracking changes in carbon emissions over time, providing insight into trends and patterns in emissions levels.



>   `iso_code`: This column typically contains the ISO 3166-1 alpha-3 country codes, which are three-letter codes assigned to countries and dependent territories by the International Organization for Standardization (ISO). These codes uniquely identify countries and are useful for aggregating or filtering data by country.





>   `country`: This column contains the names of the countries or regions for which carbon emission data is recorded. It provides the context for understanding which geographical areas are contributing to carbon emissions and allows for comparisons between different countries.





> `population`: This column represents the population of each country or region. Population size can influence the level of carbon emissions, as more populous countries tend to have higher overall emissions due to increased energy consumption, transportation needs, and industrial activities.

> `gdp`: GDP stands for Gross Domestic Product, which is a measure of the economic performance of a country or region. It represents the total value of all goods and services produced within the country's borders over a specific period, usually a year or a quarter. GDP is often correlated with carbon emissions, as higher levels of economic activity generally lead to increased energy consumption and carbon emissions.

> `co2`: This column represents the total carbon dioxide emissions (in metric tons) attributed to each country or region. Carbon dioxide is a greenhouse gas that contributes to global warming and climate change. Tracking CO2 emissions is essential for understanding a country's environmental impact and assessing progress toward emission reduction goals.

> `co2_per_capita`: This column represents the carbon dioxide emissions per capita, calculated by dividing the total CO2 emissions by the population of each country or region. It provides insight into the average emissions produced by individuals within a given population, allowing for comparisons of emission levels between countries with different population sizes.

**gdp_per_capita**

```python
# gdp_per_capita
df_full['gdp_per_capita'] = (df_full['gdp']/df_full['population'])
```

**decade**

```python
# decades
def take_decade(i):
  if i>=1750 and i<1800:
    return ((i//100)+1)
  elif i==1800:
    return (i//100)
  elif i>1800 and i<1900:
    return ((i//100)+1)
  elif i==1900:
    return (i//100)
  elif i>1900 and i<2000:
    return ((i//100)+1)
  elif i==2000:
    return (i//100)
  elif i>2000 and i<2100:
    return ((i//100)+1)
```

```python
# create columns for decade
df_full['decade'] = df_full['year'].apply(take_decade)
```

**gdp_growth_rate**

```python
# gdp_growth_rate
# Percentage Change = [(Previous Value - Current Value)/Previous Value ] * 100
```

```python
# Calculating the growth rate of GDP country-wise
df_full['gdp_growth_rate'] = df_full.groupby('country')['gdp'].pct_change() * 100
# but problem : 1st record of each country in Nan --> need to feel 0
```

```python
# Handling the first NaN value within each country group by replacing it with 0
df_full['gdp_growth_rate'] = df_full['gdp_growth_rate'].fillna(0)
```

### Create related Categorical columns from existing.

```python
df_full.describe()
```

**co2_emission_level**
- depends on co2

```python
# df_full['co2'].min() -- df_full['co2'].quantile(0.33) -> Low
# df_full['co2'].quantile(0.33) -- df_full['co2'].quantile(0.66) -> Medium
# df_full['co2'].quantile(0.66) -- df_full['co2'].max() -> High
df_full['co2_emission_level'] = pd.cut(df_full['co2'],
                                       bins=[df_full['co2'].min(), df_full['co2'].quantile(0.33), df_full['co2'].quantile(0.66)	,df_full['co2'].max()],
                                       labels=['Low', 'Medium', 'High'])
```

**development_status**
- depends on gdp_per_capita

```python
# df_full['gdp_per_capita'].min() -- df_full['gdp_per_capita'].quantile(0.33)-> Underdeveloped
# df_full['gdp_per_capita'].quantile(0.33) -- df_full['gdp_per_capita'].quantile(0.66)-> Developing
# df_full['gdp_per_capita'].quantile(0.66) -- df_full['gdp_per_capita'].max() -> Developed
df_full['development_status'] = pd.cut(df_full['gdp_per_capita'],
                                       bins=[df_full['gdp_per_capita'].min(),df_full['gdp_per_capita'].quantile(0.33), df_full['gdp_per_capita'].quantile(0.66)	,df_full['gdp_per_capita'].max()],
                                       labels=['Underdeveloped', 'Developing', 'Developed'])
```

**policy_status_need**
  *   depends on co2 and gdp

```python
threshold_co2 = df_full['co2'].quantile(0.50)
threshold_gdp = df_full['gdp'].quantile(0.50)
```

```python
df_full.loc[(df_full['co2']>threshold_co2) & (df_full['gdp']<threshold_gdp),'policy_control'] ='Improvment'
df_full.loc[~((df_full['co2']>threshold_co2) & (df_full['gdp']<threshold_gdp)),'policy_control'] ='No change'
```
